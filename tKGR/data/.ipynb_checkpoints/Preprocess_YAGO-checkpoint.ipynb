```python
entities = dict()
relations = dict()
```

```python
dataset = "YAGO"
datadir = dataset #os.path.join(DataDir, dataset)
START_YEAR = 1830
```

```python
data_type=['train','valid','test']
events = []
for dt in data_type:
    with open(os.path.join(datadir, "{}.txt".format(dt)), 'r', encoding='utf-8') as f:
        data = f.readlines()
        data_split = []
        for line in data:
            if line.endswith('\n'):
                line = line[:-1]
            event = [_.strip() for _ in line.split("\t")]
                
            try:
                start_year = int(event[3].split('-')[0])
                if start_year >= START_YEAR:
                    end_year = event[4].split('-')[0]
                    if end_year == '####':
                        end_year = start_year
                    else:
                        end_year = int(end_year)

                    for year in range(start_year, end_year+1):
#                         if year > 2020:
#                             print(event)
#                             print(start_year, end_year)
                        evt = list(map(int, event[:3])) + [year,]
                        events.append(evt)
            except:
                print(event)
            
events = sorted(events, key=lambda x: x[3])
```

```python
print("number of relations: ", len(set([evt[1] for evt in events])))
entity_idx = set([evt[0] for evt in events] + [evt[2] for evt in events])
print("number of entities: ", len(entity_idx))
print("entity index range {} ~ {}".format(min(entity_idx), max(entity_idx)))
print("number of events: ", len(events))
print("Time span: ", min([evt[3] for evt in events]), '~', max([evt[3] for evt in events]))
```

## entity id

```python
d2new_ent_id = {idx: i for i, idx in enumerate(entity_idx)}
```

```python
id2entity = {}
with open(os.path.join(datadir, "entity2id.txt"), 'r', encoding='utf-8') as f:
    data = f.readlines()
    data_split = []
    for line in data:
        if line.endswith('\n'):
            line = line[:-1]
        temp = [_.strip() for _ in line.split("\t")]
        id2entity[int(temp[1])] = temp[0]
```

```python
new_id2entity = {d2new_ent_id[k]:id2entity[k] for k in entity_idx}
```

```python
new_id2entity
```

```python
DataDir = ''
```

```python
if not os.path.exists(os.path.join(DataDir, 'YAGO{}'.format(START_YEAR))):
    os.makedirs(os.path.join(DataDir, 'YAGO{}'.format(START_YEAR)))
    
with open(os.path.join(DataDir, 'YAGO{}'.format(START_YEAR),"entity2id.txt"), 'w', encoding='utf-8') as f:
    for idx in range(len(entity_idx)):
        f.write("{}\t{}\n".format(new_id2entity[idx], idx))
```

## Training, Validation, Testing Set

```python
new_events = [[d2new_ent_id[evt[0]], evt[1], d2new_ent_id[evt[2]], evt[3]-START_YEAR] for evt in events]
#new_events = [[d2new_ent_id[evt[0]], evt[1], d2new_ent_id[evt[2]], evt[3]] for evt in events]
```

```python
num_events = len(new_events)
num_train = int(0.7*num_events)
num_val = int(0.85*num_events)
events_train = new_events[:num_train]
events_val = new_events[num_train:num_val]
events_test = new_events[num_val:]
```

```python
datadir = os.path.join(DataDir, 'YAGO{}'.format(START_YEAR))
with open(os.path.join(datadir, 'train.txt'), 'w', encoding='utf-8') as f:
    for event in events_train:
        f.write('{}\t{}\t{}\t{}\t{}\n'.format(int(event[0]), int(event[1]), int(event[2]), int(event[3]), -1))
        
with open(os.path.join(datadir, 'valid.txt'), 'w', encoding='utf-8') as f:
    for event in events_val:
        f.write('{}\t{}\t{}\t{}\t{}\n'.format(int(event[0]), int(event[1]), int(event[2]), int(event[3]), -1))
        
with open(os.path.join(datadir, 'test.txt'), 'w', encoding='utf-8') as f:
    for event in events_test:
        f.write('{}\t{}\t{}\t{}\t{}\n'.format(int(event[0]), int(event[1]), int(event[2]), int(event[3]), -1))
```
